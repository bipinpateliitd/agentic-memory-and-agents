```python
import sqlite3

conn = sqlite3.connect("students.db")
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS students_info (
    id INTEGER PRIMARY KEY,
    name TEXT,
    age INTEGER,
    marks FLOAT
)
""")

conn.commit()
```

```python
cursor.execute("INSERT INTO students_info VALUES (1, 'Rahul', 21, 85.5)")
cursor.execute("INSERT INTO students_info VALUES (2, 'Anita', 20, 92)")
cursor.execute("INSERT INTO students_info VALUES (3, 'Aman', 22, 78)")

conn.commit()
```

```python
cursor.execute("SELECT * FROM students_info")
rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
cursor.execute("SELECT name, marks FROM students_info WHERE marks > 80")
print(cursor.fetchall())
```

```python
cursor.execute("SELECT * FROM students_info ORDER BY marks DESC")
print(cursor.fetchall())
```

```python
cursor.execute("DELETE FROM students_info WHERE id = 1")
conn.commit()
```

```python
cursor.execute("SELECT COUNT(*) FROM students_info")
print("Total students:", cursor.fetchone()[0])

cursor.execute("SELECT AVG(marks) FROM students_info")
print("Average marks:", cursor.fetchone()[0])
```
